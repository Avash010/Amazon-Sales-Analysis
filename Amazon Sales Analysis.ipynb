```python
# Importing Pandas library and naming it as pd
import pandas as pd 
```

```python
# Reading the CSV file "Amazon Sales data.csv" into a Pandas DataFrame and set the first column as the index
df = pd.read_csv("Amazon Sales data.csv", index_col = 0)
df
```

```python
# To see number of rows and columns in the DataFrame.
df.shape
```

```python
# To display the first few rows of a DataFrame.
df.head()
```

```python
# To get an overview of a DataFrame.
df.info()
```

```python
# To access the column labels of a DataFrame.
df.columns
```

```python
# To remove duplicate rows based on Order ID (If any), while keeping the first occurrence of each duplicate.
df_unique = df.drop_duplicates(subset='Order ID', keep='first')
df_unique
```

```python
# To summarize descriptive statistics of the numerical columns of a DataFrame. 
df_unique.describe()
```
